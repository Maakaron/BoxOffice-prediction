```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('movies-refined-X3.csv')
data.shape
```

```python
display(data)
```

```python
data['DirectorGrossSum']=1.0
```

```python
for index, row in data.iterrows():
    director = row['Director']
    rdate = row['year']
    #query dataset
    df2 = data[(data['year']< rdate) & (data['Director']==director) ]
    rsum = max(df2['Worldwide_gross_income'].sum(),1.0)
    data.at[index, 'DirectorGrossSum']=rsum
```

```python
data['ProductionCompSum']=1.0
```

```python
data2 = data
data2['ProductionCompSum']=1.0
for index, row in data2.iterrows():
    pc = eval(row['production_companies'])
    gr = eval(row['genres'])
    pcids = []
    genres = []
    for company in pc:
        pcids.append(company['id'])
    for genre in gr:
        genres.append(genre['id'])
    data2.at[index, 'genres']=genres
    data2.at[index, 'production_companies']=pcids
display(data)
data.dtypes
```

```python
mask = data2.production_companies.apply(lambda x: len(x)>0)
mask2 = data2.genres.apply(lambda x: len(x)>0)
data2 = data2[mask]
data2 = data2[mask2]
data2.shape
```

```python
for index, row in data2.iterrows():
    if index%100==0:
        print("index:{}".format(index))
    pcs = row['production_companies']
    if len(pcs)==0: print(index)
    rdate = row['year']
    rsum = 1.0
    for company in pcs:
        mask = data2.production_companies.apply(lambda x: company in x)
        df2 = data2[(data2['year']< rdate) & (mask)]
        rsum += df2['Worldwide_gross_income'].sum()
    data2.at[index, 'ProductionCompSum']=rsum/len(pcs)
```

```python
all_genres_set = set()
for entry in data2['genres']:
    all_genres_set.update(set(entry))
all_genres_set
```

```python
for genreId in all_genres_set:
    data2['genre_'+str(genreId)]=0
data2
```

```python
# generate labels
def generate_label(df):
    conditions = [
        (df['Worldwide_gross_income'] <= 1000000),
        (df['Worldwide_gross_income'] > 1000000) & (df['Worldwide_gross_income'] <= 10000000),
        (df['Worldwide_gross_income'] > 10000000) & (df['Worldwide_gross_income'] <= 20000000),
        (df['Worldwide_gross_income'] > 20000000) & (df['Worldwide_gross_income'] <= 40000000),
        (df['Worldwide_gross_income'] > 40000000) & (df['Worldwide_gross_income'] <= 65000000),
        (df['Worldwide_gross_income'] > 65000000) & (df['Worldwide_gross_income'] <= 100000000),
        (df['Worldwide_gross_income'] > 100000000) & (df['Worldwide_gross_income'] <= 150000000),
        (df['Worldwide_gross_income'] > 150000000) & (df['Worldwide_gross_income'] <= 200000000)]
    choices=[0,1,2,3,4,5,6,7]
    df['label']=np.select(conditions, choices, default=8)
generate_label(data)
```
